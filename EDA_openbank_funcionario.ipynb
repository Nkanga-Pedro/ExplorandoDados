# Análise Descritiva e Exploratória de OPENBANK para funcionários

Os dados que foram coletados nos funcionario, resultando em 6012
registos e 14 colunas, derivadas das questões feitas durante o inquérito. Durante o préprocessamento dos dados, identificou-se que 718 funcionários não responderam
devidamente a algumas perguntas. Exceto as colunas seguintes:
• Faixa_Etaria
• Nivel_Escolaridade
• Provincia_Residencia
• Nivel_Conhecimento_Open_Banking
As respostas incompletas representam 11,94% de valores ausentes na base de dados. Para
obter melhores insights, utilizamos a técnica de Imputação Proporcional. Como já foi
mencionada acima. Após essa etapa, conseguimos ter uma base de dados limpa e sem
valores ausentes. Com essa base de dados limpa, podemos prosseguir com a análise
descritiva e exploratória.

```python
# Create the figure and subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 4))
# Get the colors from the current palette
palette = sns.color_palette()
# Create the bar chart
sns.countplot(data_clean['Faixa_Etaria'], ax=axes[0], palette=palette)
axes[0].set_xlabel('Número de Respondentes')
axes[0].set_ylabel('Faixa Etária')
axes[0].set_title('Distribuição das Faixas Etárias dos Respondentes')
# Create the pie chart
sizes = data_clean['Faixa_Etaria'].value_counts()
labels = sizes.index.to_list()
axes[1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=palette)
axes[1].axis('equal')  # Equal aspect ratio ensures a circular pie chart
axes[1].set_title('Distribuição das Faixas Etárias dos Respondentes')
# Show the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(13, 5))
palette = sns.color_palette()
# Create the bar chart
sns.countplot(data=data_clean, x='Nivel_Escolaridade', order=data_clean['Nivel_Escolaridade'].value_counts().index, palette=palette)
# Add percentage labels
total = len(data_clean)
for i in ax.patches:
    percentage = '{:.1f}%'.format(100 * i.get_height() / total)
    x = i.get_x() + i.get_width() / 2
    y = i.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')
ax.set_xlabel('Nível de Escolaridade', fontsize=12)
ax.set_ylabel('Número de Respondentes', fontsize=12)
ax.set_title('Distribuição do Nível de Escolaridade dos Respondentes', fontsize=14)
# Show the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
# Create the bar chart
sns.countplot(x='Faixa_Etaria', hue='Nivel_Escolaridade', data=data_clean, order=data_clean['Faixa_Etaria'].value_counts().index, palette='Set3')
# Add percentage labels
total = len(data_clean)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')
# Change the color of the bars
ax.patches[0].set_color('#117A65')
ax.patches[1].set_color('#A93226')
ax.patches[2].set_color('#2471A3')
ax.patches[3].set_color('#2471A3')
ax.patches[4].set_color('#2471A3')
ax.patches[5].set_color('#2471A3')
# Add title and labels
ax.set_xlabel('Faixa Etária', fontsize=12)
ax.set_ylabel('Número de Respondentes', fontsize=12)
ax.set_title('Distribuição de Faixa Etária e Nível de Escolaridade', fontsize=14)
# Show the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 10))
# Create the bar chart
sns.countplot(x='Provincia_Residencia', data=data_clean, order=data_clean['Provincia_Residencia'].value_counts().index, palette='Set3')
# Add percentage labels
total = len(data_clean)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    count = p.get_height()
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(f'{percentage}\n({count})', (x, y), ha='center', va='bottom')
# Add title and labels
ax.set_xlabel('Província de Residência', fontsize=12)
ax.set_ylabel('Número de Respondentes', fontsize=12)
ax.set_title('Distribuição de Província de Residência', fontsize=14)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
# Show the plot
plt.show()
```
