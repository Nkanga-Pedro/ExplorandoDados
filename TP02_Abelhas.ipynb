```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score
from imblearn.under_sampling import RandomUnderSampler

# configurar o estilo dos gr√°ficos com o Seaborn
sns.set_style('dark')
```

```python
# ver o balanceamento das classes
print(df_norm.res_espece.value_counts())
print("\n presen√ßa de abelha representam {:.4f}% do dataset.\n".format((df_norm[df_norm.res_espece == 1].shape[0] / df_norm.shape[0]) * 100))

# plotar gr√°fico de barras para as Classes
sns.countplot('res_espece', data=df_norm);
```

```python
# separar vari√°veis entre X e y
X = df_norm.drop('res_espece', axis=1)
y = df_norm['res_espece']
```

```python
# dividir o dataset entre treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, shuffle=True)
```

```python
# testando o modelo com essa descrepancia de dados

# instanciar e treinar um modelo de Regress√£o Log√≠stica
model = LogisticRegression()
model.fit(X_train, y_train)

# fazer as previs√µes em cima dos dados de teste
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)

# plotar a matrix de confus√£o
skplt.metrics.plot_confusion_matrix(y_test, y_pred, normalize=True)

# imprimir relat√≥rio de classifica√ß√£o
print("Relat√≥rio de Classifica√ß√£o:\n", classification_report(y_test, y_pred, digits=4))

# imprimir a acur√°cia do modelo
print("Acur√°cia: {:.4f}\n".format(accuracy_score(y_test, y_pred)))

# imprimir a √°rea sob a curva
print("AUC: {:.4f}\n".format(roc_auc_score(y_test, y_pred)))
```

```python
# usar t√©cnica under-sampling para desbalancear os dados
rus = RandomUnderSampler()
```

```python
X_res, y_res = rus.fit_resample(X_train, y_train)
```

```python
# ver o balanceamento das classes
print(pd.Series(y_res).value_counts())

# plotar a nova distribui√ß√£o de classes
sns.countplot(y_res);
```

```python
# instanciar e treinar um modelo de Regress√£o Log√≠stica
model_res = LogisticRegression()
model_res.fit(X_res, y_res)
```

```python
# fazer as previs√µes em cima dos dados de teste
y_pred_res = model_res.predict(X_test)
y_proba_res = model_res.predict_proba(X_test)
```

```python
# plotar a matrix de confus√£o
skplt.metrics.plot_confusion_matrix(y_test, y_pred_res, normalize=True)
```

```python
# imprimir relat√≥rio de classifica√ß√£o
print("Relat√≥rio de Classifica√ß√£o:\n", classification_report(y_test, y_pred_res, digits=4))
```

```python
from sklearn import metrics
import math as m
#y_pred_res
#roc_auc_score(y_test, y_pred_res)
fpr, tpr, _ = metrics.roc_curve(y_test,y_pred_res)
auc = metrics.roc_auc_score(y_test, y_pred_res)
plt.plot(fpr,tpr,label="Dado com predi√ß√£o em Logistica, AUC= "+str(m.floor(auc*100))+'%',color="darkorange")
plt.legend(loc=4)
plt.show()
```
